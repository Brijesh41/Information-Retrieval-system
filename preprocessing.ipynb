```
def ending_with(ending, words):
  return [w.replace('ending','') for w in words]
```

```
from nltk.corpus import stopwords

# You will have to download the set of stop words the first time
import nltk
nltk.download('stopwords')
```

```

mystring = "My favorite colors are blue, red, and green. also daunting black but prefered grey"
newstring = mystring.split()
```

```
print(newstring)
```

```
stopwords = set(stopwords.words('english'))
```

```
print(stopwords)
```

```
newwords = [word for word in newstring if word not in stopwords]
```

```
print(newwords)
nelist = ending_with('ing',newwords)
nelist = ending_with('ed',nelist)
```

```
print(newwords)
print(nelist)
```
